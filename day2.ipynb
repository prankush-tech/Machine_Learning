```python
iris = pd.read_csv('iris.csv')
iris.columns=['sepal-length',"sepal-width",'petal-length','petal-width','class']
iris
```

```python
np.unique(iris['class'])
```

```python
mapping_data={'Iris-setosa':0,
         'Iris-versicolor':1,
          'Iris-virginica':2}
```

```python
iris['class']= iris['class'].map(mapping_data)
iris
```

```python
le=LabelEncoder()
iris['class']=le.fit_transform(iris['class'])
iris
```
